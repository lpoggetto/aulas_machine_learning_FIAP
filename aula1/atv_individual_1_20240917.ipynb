```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Atividade 1 - Estatística

EXERCÍCIO: Estatística

* Defina categorias de altura: muito baixo, baixo, médio, alto, muito alto
* Apresente o histograma
* Calcule os quartis
* Apresente o boxplot

```python
dados_atividade = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/data/main/altura%20-%20masc.csv')
```

```python
dados_atividade.head()
```

```python
dados_atividade['altura_cat'] = pd.qcut(dados_atividade['altura_masc'], q=5, labels=['muito baixo', 'baixo', 'médio', 'alto', 'muito alto'])
```

```python
# Dados distribuidos
dados_atividade.head()
```

```python
sns.displot(dados_atividade['altura_masc'])
plt.title('Distribuição de altura')
plt.show()
```

```python
# Calculando os quartis e plotando
#dados_atividade['altura_masc'].describe()
dados_atividade['altura_masc'].quantile([0.25,0.5,0.75])
```

```python
sns.boxplot(x = dados_atividade['altura_masc'])
plt.title('Boxplot - Altura')
plt.show()
```
